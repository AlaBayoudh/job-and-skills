```python
import pandas as pd
import re
Tags_skills = pd.read_csv("data/question_tags.csv")
Tag_df_unique = Tags_skills.drop_duplicates(subset="Tag")
```

```python
print(list(Tag_df_unique.sample(50).Tag))
```

### Split tags with "-" and change special characters with space

```python
Tag_df_unique = Tag_df_unique.assign(tag_split=(lambda df: df.Tag.astype(str)
                                                .map(lambda s: re.sub("\d+", "", s))
                                                .map(lambda x: x.split("-"))))
```

```python
print(len(Tag_df_unique))
print(len(Tag_df_unique['Id'].unique()))
```

```python
from collections import Counter
from itertools import chain
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
listOftags = list(chain.from_iterable(Tag_df_unique['tag_split']))
print(listOftags[:10])
listOftags_clean = []
pat = r'[.?\-",]+'

for i in listOftags:
    i = re.sub(pat,"" ,i)
    if i not in stopWords and len(i)>1 :
        listOftags_clean.append(i)
        
res = Counter(listOftags_clean)
```

```python
dictTags = dict(res.most_common())
print(len(dictTags))
print(list(res.most_common()))
```

### We filter over tags frequent more than 10 times

```python
for k in list(dictTags):
    if dictTags[k] < 10 :
        del dictTags[k]
```
